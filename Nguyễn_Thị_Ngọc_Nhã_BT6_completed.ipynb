```python
#11. Write a Python function that takes two lists and returns True if they have at least one common member.
list_a_11 = ['a', 'b' , 'c', 1, 2, 3]
list_b_11 = ['c', 'd', 'e', 1, 2, 3]
# Using set intersection to check for common elements
common_member_exists_11 = bool(set(list_a_11) & set(list_b_11))
print(common_member_exists_11)

#12. Write a Python program to print a specified list after removing the 0th, 4th and 5th elements.
#Sample List : ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']
#Expected Output : ['Green', 'White', 'Black']
list_12 = ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']
indices_to_remove_12 = [0, 4, 5]
new_list_12 = [element for index, element in enumerate(list_12) if index not in indices_to_remove_12]
print(new_list_12)

#13. Write a Python program to generate a 3*4*6 3D array whose each element is *.
rows_13 = 3
columns_13 = 4
depth_13 = 6
array_3d_13 = [[['*' for _ in range(depth_13)] for _ in range(columns_13)] for _ in range(rows_13)]
print(array_3d_13)

#14. Write a Python program to print the numbers of a specified list after removing even numbers from it.
list_14 = [1, 3, 5, 2, 4, 6]
odd_numbers = [number for number in list_14 if number % 2 != 0]
print(odd_numbers)

#15. Write a Python program to shuffle and print a specified list.
import random
list_15 = ['a', 'b', 'c', 'd', 'e']
random.shuffle(list_15)
print(list_15)

#16. Write a Python program to generate and print a list
#of the first and last 5 elements where the values
#are square numbers between 1 and 30 (both included).
squares_16 = [x**2 for x in range(1, 31)]
first_five_16 = squares_16[:5]
last_five_16 = squares_16[-5:]
print (first_five_16)
print(last_five_16)

#17. Write a Python program to check if each number is prime in a given list of numbers.
#Return True if all numbers are prime otherwise False.
#Sample Data:
#([0, 3, 4, 7, 9]) -> False
#([3, 5, 7, 13]) -> True
#([1, 5, 3]) -> False
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

list_a_17 = [0, 3, 4, 7, 9]
new_list_a_17 = all(is_prime(num) for num in list_a_17)
print(new_list_a_17)

list_b_17 = [3, 5, 7, 13]
new_list_b_17 = all(is_prime(num) for num in list_b_17)
print(new_list_b_17)

list_c_17 = [1, 5, 3]
new_list_c_17 = all(is_prime(num) for num in list_c_17)
print(new_list_c_17)

#18. Write a Python program to generate all permutations of a list in Python.
import itertools
list_18 = [1, 2, 3, 4, 5]
permutations = list(itertools.permutations(list_18))
print (permutations)

#19. Write a Python program to calculate the difference between the two lists.
list_a_19 = [1, 2, 3, 4, 5]
list_b_19 = [4, 5, 6, 7, 8]
difference_19 = list(set(list_a_19) - set(list_b_19))
print(difference_19)

#20. Write a Python program to access the index of a list.
list_20 = ['apple', 'cherry', 'banana']
for index, value in enumerate(list_20):
  print(f"index: {index}, value: {value}")
```
