```python
import random

def play_game():
    """Main function to run the number guessing game."""

    # Initialize game variables
    win_count = 0
    loss_count = 0
    balance = 100
    cost_per_game = 5
    should_continue = True

    # Main game loop. It continues as long as the player has money and wants to play.
    while should_continue and balance >= cost_per_game:
        print(f"\n--- New Game ---")
        print(f"Current Balance: ${balance}")
        print(f"Cost of this game: ${cost_per_game}")

        balance -= cost_per_game  # Deduct money at the start of the game

        print('\nComputer thinks of a number from 1 to 100')
        comp_number = random.randint(1, 100)

        # Loop to get a valid difficulty level
        while True:
            try:
                level_input = input('Choose level [1, 2, 3]: ')
                level = int(level_input)

                # Determine the number of guesses based on the level
                if level == 1:
                    times = 10
                    print("You have 10 times to guess the number.")
                    break  # Exit this inner loop
                elif level == 2:
                    times = 5
                    print("You have 5 times to guess the number.")
                    break  # Exit this inner loop
                elif level == 3:
                    times = 3
                    print("You have 3 times to guess the number.")
                    break  # Exit this inner loop
                else:
                    print(f"'{level_input}' is not a valid choice. Please choose again.")
            except ValueError:
                print(f"'{level_input}' is not a valid number. Please choose again.")

        is_win = False

        # Loop for the actual guessing attempts
        for attempt in range(times):
            try:
                your_num = int(input(f"Enter your guessing number #{attempt + 1}: "))
                if your_num == comp_number:
                    is_win = True
                    break  # Exit this guessing loop immediately
                elif your_num < comp_number:
                    print('Too low!')
                else:
                    print('Too high!')
            except ValueError:
                print("Invalid input. Please enter a number.")

        # Check the outcome of the game
        if is_win:
            print("\nYou are a Genius!!!!")
            win_count += 1
            # You can add a reward for winning, e.g.,
            # balance += 10
        else:
            print('\nGame over!')
            loss_count += 1

        print('-------------------------')
        print(f"Total Wins: {win_count}")
        print(f"Total Losses: {loss_count}")
        print(f"Remaining Balance: ${balance}")
        print('-------------------------')

        # Check if the player has run out of money
        if balance < cost_per_game:
            print("You've run out of money! Game over.")
            should_continue = False
            break

        # Loop to get a valid choice to play again
        while True:
            cont = input("Dare you to play again [y/n]: ")
            if cont.lower() == 'n':
                should_continue = False
                break  # Exit this inner loop
            elif cont.lower() == 'y':
                break  # Exit this inner loop to continue the game
            else:
                print("Invalid input. Please choose again.")

    print("Thank you for playing!")

# Call the main function to run the game
play_game()
```
