```python
# Write a game that simulate rolling a pair of dice.
# - If the sum of two faces is greater than 5 → “Tài”
# - Otherwise → “Xỉu”
# - User ask for guessing “Tài” or “Xỉu”
# - Match the results
# - After one turn, ask user for continue playing game.
# - **** Extend the game by asking the user to enter an amount of money, then
# continue playing until the user runs out of money or the user stops playing. Statistics of results.
import random

def play_game():
  """Main function to simulate rolling a pair of dice."""

  #Get the player to enter their starting money
  try:
    money = int(input("Vui lòng nhập số tiền ban đầu: "))
  except ValueError:
    print("Số tiền không hợp lệ. Vui lòng nhập lại.")
    return
  win_count = 0
  lose_count = 0
  round_count = 0
  should_continue = True

  # Main game loop.
  #The game will continue as long as the player still have enough money or the player decides to stop playing.
  while money > 0 and should_continue:
    print(f"\nSố tiền hiện có: {money}")

    #Betting loop
    valid_bet = False
    while not valid_bet:
      try:
        bet = int(input("Nhập số tiền đặt cược: "))
        if bet > money:
          print("Không đủ tiền đặt cược. Vui lòng nhập số tiền khác.")
        elif bet <= 0:
          print("Số tiền đặt cược phải lớn hơn 0. Vui lòng nhập số tiền khác.")
        else:
          valid_bet = True #Valid input. Break the betting loop
      except ValueError:
          print("Số tiền không hợp lệ. Vui lòng nhập lại.")

      #role the dice:
      dice_1 = random.randint(1,6)
      dice_2 = random.randint(1,6)
      total = dice_1 + dice_2

      #Guessing loop
      while True:
        try:
          guess = int(input("\nDự đoán kết quả [1: Tài (>5), 2: Xỉu (<=5)]: "))
          if guess < 1 or guess > 2:
            print("Dự đoán không hợp lệ. Vui lòng nhập lại.")
          else:
            break #Valid input. Break the guessing loop
        except ValueError:
          print("Dự đoán không hợp lệ. Vui lòng nhập lại.")

      #Outcomes
      if (total >5 and guess == 1) or (total <= 5 and guess ==2):
        print("\nBạn thắng!")
        money += bet
        win_count += 1
      else:
        print("Bạn thua!")
        money -= bet
        lose_count += 1

      print(f"\nKết quả: xúc xắc 1 = {dice_1}, xúc xắc 2 = {dice_2}, tổng = {total} ")
      print(f"Số tiền hiện có: {money}")
      round_count += 1

      #Check if the player still have enough money to continue playing.
      if money <= 0:
        print("Bạn đã hết tiền. Trò chơi kết thúc.")
        break

      #Ask the player if they wants to continue the game
      while True:
        cont = input("\nBạn có muốn tiếp tục chơi không? [c: có, k: không]: ")
        if cont == 'k' or cont == 'K':
          should_continue = False
          print("Cảm ơn bạn đã chơi!")
          break
        elif cont == 'c' or cont == 'C':
          break
        else:
          print("Nội dung không hợp lệ. Vui lòng xác nhận lại.")

  #Statistic of results:
  print("\n---Thống kê kết quả---")
  print(f"Tổng số lần chơi: {round_count}")
  print(f"Tổng số lần thắng: {win_count}")
  print(f"Tổng số lần thua: {lose_count}")
  print(f"Tổng số tiền: {money}")

play_game()
```
