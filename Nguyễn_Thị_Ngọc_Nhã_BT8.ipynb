```python
#1.Write a Python function to find the maximum of three numbers.
def max_of_three(x, y, z):
    """
    Finds the maximum of three numbers.

    Args:
        x: The first number.
        y: The second number.
        z: The third number.

    Returns:
        The largest of the three numbers.
    """
    if x >= y and x >= z:
        return x
    elif y >= x and y >= z:
        return y
    else:
        return z

# Test the function with some examples
print(f"\nThe maximum of 3, 6, and 9 is: {max_of_three(3, 6, 9)}")
print(f"The maximum of 10, 5, and 2 is: {max_of_three(10, 5, 2)}")
print(f"The maximum of 1, 8, and 4 is: {max_of_three(1, 8, 4)}")
#2.Write a Python function to sum all the numbers in a list.
def sum_list_numbers(numbers):
  """
  Calculates the sum of all numbers in a list.

  Args:
    numbers: A list of numbers.

  Returns:
    The sum of all numbers in the list.
  """
  total = 0
  for number in numbers:
    total += number
  return total

# Example calls
print(f"\nSum of [1, 2, 3, 4, 5]: {sum_list_numbers([1, 2, 3, 4, 5])}")
print(f"Sum of []: {sum_list_numbers([])}")
print(f"Sum of [10]: {sum_list_numbers([10])}")
print(f"Sum of [-1, 0, 1]: {sum_list_numbers([-1, 0, 1])}")

#3.Write a Python program to reverse a string
def reverse_string(input_string):
  """Reverses a given string."""
  return input_string[::-1]

# Sample string
sample_string = "Hello, World!"
reversed_string = reverse_string(sample_string)
print(f"\nThe original string is: {sample_string}")
print(f"The reversed string is: {reversed_string}")
#4.Write a Python function to calculate the factorial of a number (a non-negative integer). The function accepts the number as an argument.
def factorial(n):
    """
    Calculates the factorial of a non-negative integer.

    Args:
        n: A non-negative integer.

    Returns:
        The factorial of n.
    """
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Test the function
print(f"\nThe factorial of 5 is: {factorial(5)}")
print(f"The factorial of 0 is: {factorial(0)}")
print(f"The factorial of 10 is: {factorial(10)}")

#5.Write a Python function that takes a number as a parameter and checks whether the number is prime or not.
def is_prime(number):
  """
  Checks if a number is prime.

  Args:
    number: An integer.

  Returns:
    True if the number is prime, False otherwise.
  """
  if number <= 1:
    return False
  for i in range(2, int(number**0.5) + 1):
    if number % i == 0:
      return False
  return True

# Test the function
print(f"\nIs 7 prime? {is_prime(7)}")
print(f"Is 10 prime? {is_prime(10)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 2 prime? {is_prime(2)}")

#6.Write a Python function to print
#6.1.all prime numbers that less than a number (enter prompt keyboard).
#6.2.the first N prime numbers
def print_primes_less_than(limit):
  """
  Prints all prime numbers less than a given limit.

  Args:
    limit: The upper limit (exclusive).
  """
  print(f"\nPrime numbers less than {limit}:")
  for number in range(2, limit):
    if is_prime(number):
      print(number, end=" ")
  print()

def print_first_n_primes(n):
  """
  Prints the first n prime numbers.

  Args:
    n: The number of prime numbers to print.
  """
  print(f"\nThe first {n} prime numbers:")
  count = 0
  number = 2
  while count < n:
    if is_prime(number):
      print(number, end=" ")
      count += 1
    number += 1
  print()

# Test the functions
print_primes_less_than(20)
print_first_n_primes(10)

#7.Write a Python function to check whether a number is "Perfect" or not. Then print all perfect number that less than 1000.
def is_perfect_number(number):
  """
  Checks if a number is a perfect number.

  Args:
    number: An integer.

  Returns:
    True if the number is a perfect number, False otherwise.
  """
  if number <= 1:
    return False
  divisors_sum = 0
  for i in range(1, number):
    if number % i == 0:
      divisors_sum += i
  return divisors_sum == number

def print_perfect_numbers_less_than(limit):
  """
  Prints all perfect numbers less than a given limit.

  Args:
    limit: The upper limit (exclusive).
  """
  print(f"\nPerfect numbers less than {limit}:")
  for number in range(2, limit):
    if is_perfect_number(number):
      print(number, end=" ")
  print()

# Test the perfect number functions
print_perfect_numbers_less_than(1000)

#8.Write a Python function to check whether a string is a pangram or not.
#(Note : Pangrams are words or sentences containing every letter of the alphabet at least once. For example : "The quick brown fox jumps over the lazy dog"
def is_pangram(text):
    """
    Checks if a string is a pangram.

    Args:
        text: The input string.

    Returns:
        True if the string is a pangram, False otherwise.
    """
    alphabet = set("abcdefghijklmnopqrstuvwxyz")
    return alphabet <= set(text.lower())

# Test the pangram function
print(f"\n'The quick brown fox jumps over the lazy dog' is a pangram: {is_pangram('The quick brown fox jumps over the lazy dog')}")
print(f"'Hello, World!' is a pangram: {is_pangram('Hello, World!')}")
```
